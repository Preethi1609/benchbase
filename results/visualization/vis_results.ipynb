```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#TODO: Add absolute path from the computer which this is being run on 
path_to_data_files = '/Users/saumyabakshi/Desktop/GT/norp/benchbase/results'
```

```python


# Read the CSV file into a DataFrame
cockroach_single_node_df = pd.read_csv(path_to_data_files + '/cockroach/results_tpcc_cockroach_1h_3_and_1_node/tpcc_2024-04-21_11-59-28.results.csv')
cockroach_three_nodes_df = pd.read_csv(path_to_data_files + '/cockroach/results_tpcc_cockroach_1h_3_and_1_node/tpcc_2024-04-21_10-25-41.results.csv')

# Plotting throughput
plt.figure(figsize=(10, 5))

# Plot throughput for three nodes
plt.plot(cockroach_three_nodes_df['Time (seconds)'], cockroach_three_nodes_df['Throughput (requests/second)'],
         color='blue', linestyle='-', label='Throughput - Three Nodes')

# Plot throughput for single node
plt.plot(cockroach_single_node_df['Time (seconds)'], cockroach_single_node_df['Throughput (requests/second)'],
         color='red', linestyle='-', label='Throughput - Single Node')

plt.xlabel('Time (seconds)')
plt.ylabel('Throughput (requests/second)')
plt.title('Throughput Comparison')
plt.legend()
plt.grid(True)
plt.show()

# Plotting latency
plt.figure(figsize=(10, 5))

# Plot average latency for three nodes
plt.plot(cockroach_three_nodes_df['Time (seconds)'], cockroach_three_nodes_df['Average Latency (millisecond)'],
         color='blue', linestyle='-', label='Average Latency - Three Nodes')

# Plot average latency for single node
plt.plot(cockroach_single_node_df['Time (seconds)'], cockroach_single_node_df['Average Latency (millisecond)'],
         color='red', linestyle='-', label='Average Latency - Single Node')

plt.xlabel('Time (seconds)')
plt.ylabel('Latency (millisecond)')
plt.title('Latency Comparison')
plt.legend()
plt.grid(True)
plt.show()
```
